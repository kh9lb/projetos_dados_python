```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sistema de Grids

# Cor Fundo de Fundo
cor_Fundo = '#f5f5f5'

# Criar o sistema de Grids
grid_Graficos = sns.FacetGrid( base_dados, col='Territorialidades', hue='Territorialidades', col_wrap=4 )

# Adicionar grafico linhas em cada gráfico
grid_Graficos = grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

# Adiconar uma sombra + Ajuste do titulo
grid_Graficos = grid_Graficos.map( plt.fill_between,  'Ano', 'PIB per capita', alpha=0.5).set_titles('{col_name} Territorialidades')

# Filtrar o titulo
grid_Graficos = grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
grid_Graficos = grid_Graficos.fig.suptitle(
    'Evolução da Renda per capita por Estado \n Elaborado por: Khalil',
    fontsize=18
)

# Ajustando posição do subtitulo
plt.subplots_adjust( top=0.92 )
```
