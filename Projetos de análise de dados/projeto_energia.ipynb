```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base_dados = pd.read_excel(r'C:\Users\win10\Desktop\curso_python\projetos dados\Dados_Empresas_Energia.xlsx')

base_dados.head()
```

```python
#Definir a data como index

base_dados.set_index('Data', inplace=True)
```

```python
#Criando o grafico

#Cor de fundo do grafico
sns.set_style('darkgrid')

#Tamanho do grafico
plt.figure(figsize=(16,7))

#Titulo e legendas do grafico
plt.title('Análise das ações de empresas de energia', loc='left', fontsize=18)
plt.xlabel('Período',fontsize=14)
plt.ylabel('Preço Fechamento(R$)', fontsize=14)

#Plot da petrobras e na segunda linha, esta inserindo o nome petrobras na ultima data e ultimo valor disponivel na base
plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a', linewidth=3)
plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large')

#Inserindo as colunas das outras empresas dentro de uma variavel chamada coluna
for coluna in base_dados.columns[1:]:
    plt.plot(base_dados.index, base_dados[coluna], color='gray', linewidth=1);
    plt.text(base_dados.index[-1], base_dados[coluna].tail(1), coluna, color='gray', size='large')
```
