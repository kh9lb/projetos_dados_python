```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_csv(r'C:\Users\win10\Desktop\curso_python\projetos dados\Salary_Data.csv')
```

```python
base_dados.head()
```

```python
#Renomear as colunas
base_dados.rename( columns={
    'YearsExperience' : 'XP',
    'Salary' : 'Renda'
}, inplace=True
)
```

```python
#Verificando tamanho da base
base_dados.shape
```

```python
#Verificando Campos nulos
base_dados.isnull().sum()
```

```python
#Estatistica da base
base_dados.describe()
```

```python
#Análise da renda 
plt.figure(figsize=(10,6))
plt.title('Análise da Renda')
sns.kdeplot(base_dados['Renda'], color='green');
```

```python
#Análise da Experiencia
plt.figure(figsize=(10,6))
plt.title('Análise da XP')
sns.kdeplot(base_dados['XP'], color='blue');
```

```python
#Distribuição
plt.figure(figsize=(10,6))
plt.title('Distribuição')
sns.regplot(data=base_dados, x='Renda', y='XP');
```
